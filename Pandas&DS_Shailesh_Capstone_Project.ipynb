#### 4. Please remove special characters from the name column.

```python
import string # importing string library to get the list of alphabets

alph = list(string.ascii_letters) # creating a list of alphabets with lowercase and uppercase
print(alph)
```

```python
specChar = set() # initializing a set to store unique special characters
for name in ds.name: #iterating over the name column
  for lett in str(name): # iterating over each letter of the name 
    if lett not in alph and lett != ' ' and not lett.isnumeric(): specChar.add(lett) # finding the special characters by checking if the letter is not present in alphabets list

print(specChar)
```

```python
count_of_specChar = 0 # initializing a counter to count the number of special characters
for name in ds.name:
  for lett in str(name): 
    if lett in specChar: count_of_specChar += 1

print(count_of_specChar)
```

```python
ds['clean_name'] = ds.name # creating a new column 'clean_name' where the name with no special characters is stored
ds
```

```python
for i in specChar: # iterating over the special characters set
  ds['clean_name'] = ds['clean_name'].str.replace(i,'') # every special character is replaced for every iteration

ds # checking the clean_name column 
```

```python
count_of_specChar = 0
for name in ds.clean_name:
  for lett in str(name): 
    if lett in specChar: count_of_specChar += 1

print(count_of_specChar)
```

After replacing all the special characters the count is 0 which was perviously 2015
